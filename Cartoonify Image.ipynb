## Read an image

```python
import numpy as np
import cv2

img = cv2.imread('image.jpg')
```

## Plot the image

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.imshow(img)
plt.show()
```

## Convert BGR image to RGB image

```python
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

## Smoothening the image

```python
num_iter = 5
for _ in range(num_iter):
    image= cv2.bilateralFilter(image, d=9, sigmaColor=9, sigmaSpace=7)
```

```python
plt.imshow(image)
plt.show()
```

# Generate edge line

## Convert RGB to grayscale image

```python
img_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
img_blur = cv2.medianBlur(img_gray, 7)
```

```python
plt.imshow(img_gray)
plt.show()
```

## Retrieving the edges of the image

```python
img_edge = cv2.adaptiveThreshold(img_blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 7, 2)
```

```python
plt.imshow(img_edge)
plt.show()
```

## Convert grayscale to RGB image

```python
img_edge = cv2.cvtColor(img_edge, cv2.COLOR_GRAY2RGB)
```

```python
plt.imshow(img_edge)
plt.show()
```

## Giving a Cartoon Effect

```python
array = cv2.bitwise_and(image, img_edge)
```

```python
plt.figure(figsize= (10,10))
plt.imshow(array)

plt.show()
```
